```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot
import os

def formatHeaders(header):
    return header.lower().replace("\n", "_").replace(" ", "_").replace("-", "_").strip()
    
    

data_path = "data/"
data = []
for file in os.listdir(data_path):
    raw = pd.read_excel(data_path + file, skiprows=4)
    raw.columns = [formatHeaders(column) for column in raw.columns]
    data.append(raw)


data = pd.concat(data)
#data.rename(columns={'apart_ment_number':'apartment_number'}, inplace=True)
#print(data.columns)
data
```

```python
# filter out transactions with zero-qty & zero sales price
filtered_data = data[(data.total_units > 0) & (data.sale_price > 0)]

def get_district_by_borough(borough):
    if borough == 1:
        return "Manhattan"
    elif borough == 2:
        return "Bronx"
    elif borough == 3:
        return "Brooklyn"
    elif borough == 4:
        return "Queens"
    elif borough == 5:
        return "Staten Island"

filtered_data.insert(1, "district", [get_district_by_borough(a) for a in filtered_data.borough])

filtered_data
```

```python
sales_data = filtered_data.groupby("district", as_index = False).sum()
sns.barplot(data=sales_data, x=sales_data.district, y=sales_data.sale_price)
```

```python
count_data = filtered_data.groupby("district", as_index = False).count()
sns.barplot(data=count_data, x=count_data.district, y=count_data.sale_price)
```

```python
filtered_data['month'] = pd.DatetimeIndex(filtered_data.sale_date).month
month_data = filtered_data.groupby(["month"], as_index = False).count()
month_data
sns.barplot(data=month_data, x=month_data.month, y=month_data.sale_price)
```

```python
month_dist_data = filtered_data.groupby(["month", "district"], as_index = False).count()
month_dist_data
sns.barplot(data=month_dist_data, x=month_dist_data.month, y=month_dist_data.sale_price, hue=month_dist_data.district)
```
